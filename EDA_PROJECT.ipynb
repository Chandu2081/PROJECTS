```python
company_names=[]
employees = []
old=[]
head_office=[]
reviews=[]
type_company=[]
sector=[]
highly_rated=[]
for i in range(1,15):
    #print("""https://www.ambitionbox.com/list-of-companies?sortBy=popular&tags=startup&page="""+str(i))
    url = "https://www.ambitionbox.com/list-of-companies?sortBy=popular&tags=startup&page="+str(i)
    page = requests.get(url,headers = {'User-Agent':'Mozilla/5.0'})
    soup = BeautifulSoup(page.text)
    
    for i in soup.find_all('h2',class_="companyCardWrapper__companyName"):
        company_names.append(i.text.split()[0])
        
        
    for i in soup.find_all("span",class_="companyCardWrapper__interLinking"):
        employees.append(i.text.split("|")[1])
        
    for i in soup.find_all("span",class_="companyCardWrapper__interLinking"):
        old.append(i.text.split("|")[-2])
        
        
    for i in soup.find_all("span",class_="companyCardWrapper__interLinking"):
        head_office.append(i.text.split("|")[-1].split("+")[0])
        
        
    for i in soup.find_all("span",class_="companyCardWrapper__companyRatingValue"):
        reviews.append(i.text)
        
        
    for i in soup.find_all("span",class_="companyCardWrapper__interLinking"):
        type_company.append(i.text.split("|")[-3])
      
    
    for i in soup.find_all("span",class_="companyCardWrapper__interLinking"):
        sector.append(i.text.split("|")[0].split()[0])   
        
    container=soup.find_all('div',class_='companyCardWrapper__secondaryInformation')
    for i in container:
        a=i.find("span",class_="companyCardWrapper__ratingValues")
        if a:
            highly_rated.append(i.text.split()[0])
        else:
            highly_rated.append(np.nan)
```
